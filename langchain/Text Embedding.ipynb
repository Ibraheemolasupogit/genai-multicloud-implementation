# Example

```python
embed_king = embeddings.embed_query("king")
embed_man = embeddings.embed_query("man")
embed_woman = embeddings.embed_query("woman")
embed_queen = embeddings.embed_query("queen")
```

```python
import numpy as np
```

```python
vec_king = np.array(embed_king)
vec_man = np.array(embed_man)
vec_woman = np.array(embed_woman)
vec_queen = np.array(embed_queen)
```

```python
vec_analogy = vec_king - vec_man + vec_woman
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
similarity_score = cosine_similarity(vec_analogy, vec_queen)
```

```python
similarity_score
```

```python
embed_prince = embeddings.embed_query("prince")
vec_prince = np.array(embed_prince)
similarity_prince = cosine_similarity(vec_analogy, vec_prince)
print(f"Similarity with 'prince': {similarity_prince:.4f}")
```
